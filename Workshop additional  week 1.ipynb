```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#ead some data into a pandas Data Frame
df = pd.read_csv('iris.csv',sep=",",header=0)
```

```python
#view the first rows. 
df.head()
```

```python
#explore our Data Frame giving summary statistics
df.describe()
```

```python
#select a specific column from the Data Frame
df['SepalLengthCm']
```

```python
# we can use a histogram to see the distribution of one numerical variable
plt.hist(df['SepalLengthCm'])
plt.xlabel('sepal length')
plt.ylabel('Number of observations')
plt.title('Sepal Length Distribution')
plt.show
```

```python
# we can use a scatterplot to see the relationship between two numerical variables
plt.scatter(df['SepalWidthCm'], df['SepalLengthCm'])
plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length')
plt.title('Scatter plot of sepal width against Sepal Length')
plt.show()
```

```python
# we can also use a boxplot to summarize the distributions of numerical variables
plt.boxplot(x=[df['SepalWidthCm'],df['SepalLengthCm'],df['PetalWidthCm'],df['PetalLengthCm']])
plt.xticks([1,2,3,4],['SepalWidthCm','SepalLengthCm','PetalWidthCm','PetalLengthCm'])
plt.show()
```

```python
# to understand how frequently each category occurs. 
species_counts = pd.DataFrame(df['Species'].value_counts())
species_counts.head()
```

```python
#create a bar plot of this data using the matplotlib bar function
plt.bar(x=species_counts.index,height=species_counts['Species'])
plt.xlabel('Species')
plt.ylabel('Number of Observations in Data')
plt.title('Barplot of Number of Observations for each Species')
plt.show()
```
